## Data Cleaning

```python
#@title Memisahkan data kategorikal dan numerik
drop_col = ['id','student']
cat_cols = ['Parental_Involvement', 'Access_to_Resources', 'Extracurricular_Activities', 'Motivation_Level', 'Internet_Access', 'Family_Income', 'Teacher_Quality', 'School_Type', 'Peer_Influence', 'Learning_Disabilities','Parental_Education_Level', 'Distance_from_Home', 'Gender' ]
num_cols = [i for i in df.columns if i not in (drop_col + cat_cols)]
```

```python
#@title Handling Outlier

# Fungsi untuk menghitung persentase outlier
def hitung_persentase_outlier(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    batas_bawah = Q1 - 1.5 * IQR
    batas_atas = Q3 + 1.5 * IQR
    jumlah_outlier = ((data < batas_bawah) | (data > batas_atas)).sum()
    persentase_outlier = jumlah_outlier / len(data) * 100
    return persentase_outlier

# Menghitung persentase outlier pada setiap kolom numerik
persentase_outlier = df[num_cols].apply(hitung_persentase_outlier)

print(persentase_outlier)
```

##Analisis EDA


```python
#@title Berapa Banyak Siswa yang Dapat Mengakses Internet

plt.figure(figsize=(3, 3))
sns.countplot(x='Internet_Access', data=df)
plt.title('Jumlah siswa yang mengakses internet')
plt.show()
```

```python
#@title Distribusi Jam Belajar

# Membuat salinan DataFrame
df_vis = df.copy()

# Fungsi untuk mengelompokkan Hours Studied
def jam_belajar(Hours_Studied):
    if 0 <= Hours_Studied <= 10:
        return 'jarang belajar'
    elif 11 <= Hours_Studied<= 20:
        return 'cukup belajar'
    elif 21 <= Hours_Studied <= 30:
        return 'Sering belajar'
    else:
        return 'Sangat sering belajar'

# Membuat kolom baru 'Hours Studied'
df_vis['Kategori_Jam_Belajar'] = df_vis['Hours_Studied'].apply(jam_belajar)

# Membuat bar chart
import plotly.express as px
fig = px.histogram(df_vis, x="Kategori_Jam_Belajar", title='Jam Belajar Siswa', width=400, height=400)

fig.show()
```

```python
#@title Distribusi Nilai Ujian

# Membuat salinan DataFrame
df_vis = df.copy()

# Fungsi untuk mengelompokkan Exam_Score
def Kategori_nilai_Ujian(Exam_Score):
    if 0 <= Exam_Score <= 40:
        return 'kurang memuaskan'
    elif 41 <= Exam_Score<= 60:
        return 'cukup memuaskan'
    elif 61 <= Exam_Score <= 80:
        return 'memuaskan'
    else:
        return 'sangat memuaskan'

# Membuat kolom baru 'Exam Score'
df_vis['Kategori_nilai_Ujian'] = df_vis['Exam_Score'].apply(Kategori_nilai_Ujian)

# Membuat bar chart
import plotly.express as px
fig = px.histogram(df_vis, x='Kategori_nilai_Ujian', title='Nilai Ujian Siswa', width=400, height=400)

fig.show()
```

```python
#@title Korelasi Antara Internet Access, Hours Studied, Exam Score
# Mengubah kolom 'Internet_Access' menjadi numerik (0 untuk 'No', 1 untuk 'Yes').
df['Internet_Access_Numeric'] = pd.factorize(df['Internet_Access'])[0]

plt.figure(figsize=(10, 8))

sns.heatmap(df[['Internet_Access_Numeric', 'Hours_Studied', 'Exam_Score']].corr(), annot=True, cmap='coolwarm')
plt.title('Korelasi Antara Internet_Access,Hours_Studied, Exam_Score')
plt.show()
```

```python
#@title Hubungan kehadiran dengan Nilai Ujian
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Attendance', y='Exam_Score', data=df, legend=False)
plt.title('Hubungan kehadiran dengan Nilai Ujian')
plt.xlabel('Attendance')
plt.ylabel('Exam_Score')
plt.show()
```

```python
#@title Kategori Nilai Ujian Berdarkan Gender
plt.figure(figsize=(14, 6))
sns.countplot(data=df,x="Exam_Score",hue="Gender",palette="magma")
plt.title("Kategori Nilai Ujian Berdarkan Gender")
plt.show()
```

```python
#@title Motivation Level Belajar Siswa

sns.displot(data=df,y='Motivation_Level')
plt.title("Motivation Level Siswa")
plt.show()
```

```python
#@title Distribusi Sleep Hours dan Distribusi Hours Studied
plt.figure(figsize=(14, 6))

# Distribusi Sleep_Hours
plt.subplot(1, 2, 1)
sns.histplot(df['Sleep_Hours'], kde=True, bins=20, color='blue')
plt.title('Distribusi Sleep_Hours')

# Distribusi Hours_Studied
plt.subplot(1, 2, 2)
sns.histplot(df['Hours_Studied'], kde=True, bins=20, color='pink')
plt.title('Distribusi Hours_Studied')

plt.show()
```

```python
#@title Korelasi Antara Attendance, Sleep Hours, Exam Score
plt.figure(figsize=(10, 8))
sns.heatmap(df[['Attendance', 'Sleep_Hours', 'Exam_Score']].corr(), annot=True, cmap='coolwarm')
plt.title('Korelasi Korelasi Antara Attendance, Sleep_Hours, Exam_Score')
plt.show()
```
